```python
import pandas as pd
```

```python
beg_normal_dict = {
    "avg_flesch_reading_ease": avg_flesch_reading_ease_bn,
    "avg_flesch_kincaid_grade": avg_flesch_kincaid_grade_bn,
    "avg_smog_index": avg_smog_index_bn
    }


expert_normal_dict = {
    "avg_flesch_reading_ease": avg_flesch_reading_ease_en,
    "avg_flesch_kincaid_grade": avg_flesch_kincaid_grade_en,
    "avg_smog_index": avg_smog_index_en
    }

beg_full_dict = {
    "avg_flesch_reading_ease": avg_flesch_reading_ease_bf,
    "avg_flesch_kincaid_grade": avg_flesch_kincaid_grade_bf,
    "avg_smog_index": avg_smog_index_bf
    }

expert_full_dict = {
    "avg_flesch_reading_ease": avg_flesch_reading_ease_ef,
    "avg_flesch_kincaid_grade": avg_flesch_kincaid_grade_ef,
    "avg_smog_index": avg_smog_index_ef
    }
```

```python
beg_normal_dict = {"avg_flesch_reading_ease": 31.23,
                   "avg_flesch_kincaid_grade": 12.04}
expert_normal_dict = {"avg_flesch_reading_ease": 8.60,
                     "avg_flesch_kincaid_grade": 16.78}
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Only FRE and FKGL
metrics = ["avg_flesch_reading_ease", "avg_flesch_kincaid_grade", "avg_smog_index"]
pretty_names = ["Flesch Reading Ease", "Flesch-Kincaid Grade", "SMOG Index"]

# Values grouped: [(BN, EN), (BF, EF)] or (BN, EN) for your version
groups = [
    (beg_normal_dict["avg_flesch_reading_ease"], expert_normal_dict["avg_flesch_reading_ease"]),
    (beg_normal_dict["avg_flesch_kincaid_grade"], expert_normal_dict["avg_flesch_kincaid_grade"]),
    (beg_normal_dict["avg_smog_index"], expert_normal_dict["avg_smog_index"])
]

# Colors
colors = ["#2b83ba", "#f28e2b"]  # Beginner, Expert

# X axis setup
x = np.arange(len(metrics))  # 2 groups
width = 0.35

plt.figure(figsize=(10, 6))

# Plot beginner
plt.bar(x - width/2, [g[0] for g in groups], width, label="Beginner", color=colors[0])
# Plot expert
plt.bar(x + width/2, [g[1] for g in groups], width, label="Expert", color=colors[1])

# Labeling
plt.xticks(x, pretty_names, fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.title("Readability Comparison — Beginner vs Expert - With Ancestor Selection", fontsize=14)
plt.legend(fontsize=12)

# Annotate values
for i, (bn, en) in enumerate(groups):
    plt.text(i - width/2, bn + 0.02 * max(bn, en), f"{bn:.2f}", ha='center', fontsize=10)
    plt.text(i + width/2, en + 0.02 * max(bn, en), f"{en:.2f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
